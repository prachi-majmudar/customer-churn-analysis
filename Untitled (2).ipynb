```python
##applying chi-square test
```

```python
from scipy.stats import chi2_contingency
```

```python

contingency = pd.crosstab(df['Churn'], df['Contract'])
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"Chi-Square p-value for Churn vs Contract: {p}")
```

```python
##feature engineering
```

```python
df['ChargeRatio'] = df['MonthlyCharges'] / (df['TotalCharges'] + 1)
```

```python
df['TenureGroup'] = pd.cut(df['tenure'], bins=[0, 12, 24, 48, 72], labels=['0â€“12', '13â€“24', '25â€“48', '49â€“72'])
```

```python
sns.countplot(x='TenureGroup', hue='Churn', data=df)
plt.title("Churn by Tenure Group")
plt.show()
```

```python
##k-means clustering
```

```python
X = df[['MonthlyCharges', 'TotalCharges']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
sns.scatterplot(data=df, x='MonthlyCharges', y='TotalCharges', hue='Cluster', palette='Set1')
plt.title("Customer Segmentation by Charges")
plt.legend(title='Cluster')
plt.show()
```
